```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
import time
import pymongo
import datetime

import nltk

#sean's tokenization program
import tokenization as token

conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Declare the database
comic_db = client.comic_db

# Declare the collections
dc_collection = comic_db.dc_collection
marvel_collection = comic_db.marvel_collection

# Import the 538 Datasets
dcDF = pd.read_csv("dc-wikia-data.csv")
marvelDF = pd.read_csv("marvel-wikia-data.csv")
```

```python
## DC LOOP
index = 0

for character in dcDF['urlslug']:
    url = 'https://dc.fandom.com/wiki/' + character[8:]
    print(url)
    
    response = requests.get(url)
    results = bs(response.text, 'html')

    paragraphs = results.find_all('p')
    
    #create continuous text string 
    full_text = ""
    for r in paragraphs:
        full_text = r.text + full_text

    #Addressing new line characters
    full_text = full_text.replace("\n"," ")

    #Addressing multiple spaces
    full_text = full_text.replace("\s+"," ")
    
    #insert all into mongo database
    post = {
        'page_id': int(dcDF['page_id'][index]),
        'name': dcDF['name'][index],
        'urlslug': dcDF['urlslug'][index],
        'identity_type': dcDF['ID'][index],
        'alignment': dcDF['ALIGN'][index],
        'eye_color': dcDF['EYE'][index],
        'hair_color': dcDF['HAIR'][index],
        'sex': dcDF['SEX'][index],
        'gsm': dcDF['GSM'][index],
        'alive': dcDF['ALIVE'][index],
        'appearances': dcDF['APPEARANCES'][index],
        'first_appearance': dcDF['FIRST APPEARANCE'][index],
        'year': dcDF['YEAR'][index],
        'history_text': full_text,
        'scrape_date':datetime.datetime.utcnow()
    }
    
    dc_collection.insert_one(post)
    
    index += 1
    time.sleep(5)
    
    
```
